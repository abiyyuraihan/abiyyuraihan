```python
import numpy as np
import matplotlib.pyplot as plt

# Monthly bag production data
produksi_tas = [1863, 1614, 2570, 1685, 2101, 1811, 2457, 2171, 2134, 2502, 2358, 2399, 2048, 2523, 2086, 2391, 2150, 2340, 3129, 2277, 2964, 2997, 2747,
                2862, 3405, 2677, 2749, 2755, 2963, 3161, 3623, 2768, 3141, 3439, 3601, 3531, 3477, 3376, 4027, 3175, 3274, 3334, 3964, 3649, 3502, 3688,
                3657, 4422, 4197, 4441, 4736, 4521, 4485, 4644, 5036, 4876, 4789, 4544, 4975, 5211, 4880, 4933, 5079, 5339, 5232, 5520, 5714, 5260, 6110,
                5334, 5988, 6235, 6365, 6266, 6345, 6118, 6497, 6278, 6638, 6590, 6271, 7246, 6584, 6594, 7092, 7326, 7409, 7976, 7959, 8012, 8195, 8008,
                8313, 7791, 8368, 8933, 8756, 8613, 8705, 9098, 8769, 9544, 9050, 9186, 10012, 9685, 9966, 10048, 10244, 10740, 10318, 10393, 10986, 10635,
                10731, 11749, 11849, 12123, 12274, 11666, 11960, 12629, 12915, 13051, 13387, 13309, 13732, 13162, 13644, 13808, 14101, 13992, 15191, 15018,
                14917, 15046, 15556, 15893, 16388, 16782, 16716, 17033, 16896, 17689]

# Exponential function
def fungsi_eksponensial(x, a, b):
    hasil = a * np.exp(b * x)
    return hasil

# Manually performing exponential fitting
def fit_eksponensial(data_x, data_y):
    # Calculate the logarithm of the production data
    log_data_y = np.log(data_y)

    # Use the matrix method to find the parameters a and b
    matriks_X = np.vstack((data_x, np.ones(len(data_x)))).T
    matriks_Y = log_data_y

    # Calculate the coefficients using least squares method
    koefisien = np.linalg.lstsq(matriks_X, matriks_Y, rcond=None)[0]
    b = koefisien[0]
    a = np.exp(koefisien[1])

    # Return the values of a and b
    return a, b

# Calculate parameters with the given function
a, b = fit_eksponensial(np.arange(1, len(produksi_tas) + 1), np.array(produksi_tas))

# Generate predictions for the plot
nilai_x = np.arange(1, len(produksi_tas) + 31)  # 30 months ahead for predictions
nilai_y = fungsi_eksponensial(nilai_x, a, b)

# Visualize original data and exponential model
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(1, len(produksi_tas) + 1), produksi_tas, color='red', label='Production Data')
plt.plot(nilai_x, nilai_y, color='green', linestyle='-', linewidth=2, label='Exponential Regression')
plt.title('Bag Production Trend from January 2018 to December 2023 (Exponential Regression)')
plt.xlabel('Month (M)')
plt.ylabel('Production Quantity')
plt.legend()
plt.grid(True)
plt.show()

# Display exponential regression parameters
print("Exponential Regression Parameters (a, b):", a, b)
```
